# 2.Are there any missing values in the dataset? Identify any missing values and describe how they may impact your analysis.
In this dataset there are some errors and we have to do some data cleaning. There are some NAN values, non numeric data,duplicates value.


```python
#to check the numeric values in review column
df['Reviews'].str.isnumeric().sum()
```

```python
#to filter the non-numeric values
df[~df['Reviews'].str.isnumeric()]
```

```python
##Replacing the values
df['Reviews'] = df['Reviews'].str.replace('3.0M','3000000')
```

```python
#converting the values into integer type
df['Reviews'] = df['Reviews'].astype('int')
```

```python
#to drop the row index
df.drop(index=10472,inplace=True)
```

```python
df.info()
```

```python
#to display the unique value of size column
df['Size'].unique()
```

```python
#to convert into numeric value
def convert_size(size):
  if 'M' in size:
    return int(float(size.replace('M',''))*1000)
  elif 'k' in size:
    return float(size.replace('k',''))
  elif 'Varies with device' in size:
    return np.nan
  else:
    return np.nan
```

```python
#convert size
df['Size']=df['Size'].apply(convert_size)
```

```python
df.info()
```

```python
df['Size'].unique()
```

```python
df['Installs'].unique()
```

```python
df['Price'].unique()
```

```python
cols = ['Price']
chars_to_remove=['+',',','$']
for i in cols:
  for j in chars_to_remove:
    df[i] =df[i].str.replace(j,' ')
```

```python
cols = ['Installs']
chars_to_remove=['+',',','$']
for i in cols:
  for j in chars_to_remove:
    df[i] =df[i].str.replace(j,' ')
```

```python
df['Price'].unique()
```

```python
df['Installs'].unique()
```

```python
#as there is space in installs values so we could not convert to string so first we have to remove the space
cols = ['Installs']
chars_to_remove=[' ']
for i in cols:
  for j in chars_to_remove:
    df[i] =df[i].str.replace(j,'')
```

```python
df['Installs'].unique()
```

```python
#after removing the space now we can convert installs and price column into integer type.
df['Installs']=df['Installs'].astype('int')
df['Price'] = df['Price'].astype('float')
```

```python
df.info()
```

```python
#so for the visualization , we need numeric data aso we have to convert date into day,month and year
df['Last Updated'] =pd.to_datetime(df['Last Updated'])
df['day'] =df['Last Updated'].dt.day
df['month'] =df['Last Updated'].dt.month
df['year'] =df['Last Updated'].dt.year
```

```python
df
```

```python
#removing the duplicate item in app column
df.drop_duplicates(subset=['App'],keep='last',inplace=True)
df
```

## 3.What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python

plt.figure(figsize=(20,10))
#sns.boxplot(x=df['Category'],y=df['Rating'])
#plt.xticks(rotation=90)
#plt.show()


sns.scatterplot(x=df['Category'], y=df['Rating'], alpha=0.5)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Adding labels and title
plt.title('Scatter Plot of Rating by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
```

## 4.How does the rating of apps vary by category? Create a boxplot to compare the ratings of different app categories.

```python
#boxplot between rating and category
plt.figure(figsize=(10,20))
sns.boxplot(x=df['Category'],y=df['Rating'])
plt.xticks(rotation=90)
plt.show
```

```python
#top 10 apps with largest size
df_app_size = df['Size'].value_counts().sort_values(ascending=False).to_frame()


plt.figure(figsize=(12, 6))
df_app_size.head(10).plot(kind='bar', color='skyblue', edgecolor='black')

# Adding labels and title
plt.title('Top 10 Sizes - Bar Plot')
plt.xlabel('Size')
plt.ylabel('Count')
```

## 6.What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
#top 10 apps with largest no. of installs
df_top10 = df.sort_values(by='Installs', ascending=False).head(10)
sns.barplot(x=df_top10['App'], y=df_top10['Installs'], palette='Set2')
plt.xticks(rotation=90)
plt.xlabel('App Name')
plt.ylabel('Number of Installs')
plt.title('Top 10 Apps with Largest Number of Installs')
plt.show()

```
